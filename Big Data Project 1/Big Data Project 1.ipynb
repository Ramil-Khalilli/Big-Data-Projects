# Question 1

 Every day at certain airports, people’s body temperatures are recorded in order to identify
potentially ill people and alert them in important cases.

You are provided with airport data containing the name of the airport, the date and temperature
of the people who were at the airport on that day.

According to that data, write the MapReduce program to find out on which days high
temperatures were detected, thus identifying potentially ill people.

Output: date:count

```python
!pip install mrjob
```

```python
import pandas as pd
```

```python
with open('/kaggle/input/body-temps/airportsensor.txt', 'r') as f:
    lines = f.readlines()
lines
```

```python
%%file Question1.py

from mrjob.job import MRJob
class high_temp_count(MRJob):

    def mapper(self, _, line):
      date = line.split(',')[1].split()[0]
      degree = line.split(',')[1].split()[1]

      if float(degree) >= 37.0:
        yield date, 1


    def reducer(self, key, values):
        yield key, sum(values)


if __name__ == '__main__':
    high_temp_count.run()
```

2. You have been given cards.csv data. Write a MapReduce program which returns the number of

```python
%%file question2.py

from mrjob.job import MRJob
class suit_count(MRJob):
    
    def mapper(self, _, line):
      card_name = line.split(',')[1]
      yield card_name, 1

    def reducer(self, key, values):
      yield key, sum(values)


if __name__ == '__main__':
    suit_count.run()
```

# Question 3

3. You have been given 'Life_of_PI_good_review.txt' data. According to that data write a MapReduce program which finds word frequencies for all words.

```python
%%file Question3.py

from mrjob.job import MRJob
import re

class word_counter(MRJob):
    
    def mapper(self, _, line):
        words_list = re.findall(r'[a-z\']+', line.lower())

        for i in words_list:
          yield i, 1

    def reducer(self, key, value):
        yield key, sum(value)


if __name__ == '__main__':
    word_counter.run()
```
